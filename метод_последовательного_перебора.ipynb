```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
  return math.log(x)-5*math.cos(x)

a=2
b=11
eps=0.00001
h = 0.01 #шаг
```

```python
ff = np.vectorize(f)
x = np.arange (a,b,0.1)
y = ff(x)
plt.plot(x,y)
```

```python
def seq_enum(x0,h,eps,f):
  x1=x0
  y0=10
  y1=f(x1)
  while(abs(h)>eps/4):
    while(y1<y0):
      x0=x1
      y0=y1
      x1=x0+h
      y1=f(x1)
    # повтор спуска с меньшим шагом в обратную сторону
    h=-h/4 
    while(y1<y0):
      x0=x1
      y0=y1
      x1=x0+h
      y1=f(x1)
  return x0
```

```python
seq_enum(5,h,eps,f)
```
