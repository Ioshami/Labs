```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return math.log(x)-5*math.cos(x)

a=1.
b=8.
eps=0.00001
```

Вид функции

```python
ff = np.vectorize(f)
x = np.arange (a,b,0.1)
y = ff(x)
plt.plot(x,y)
```

x0>x1 писать

```python

def secantMethod(f,x0,x1):
#первый корень
  #x0 = (a + b) / 2
  #x1=a
#второй корень
  #x0=6
  #x1=4
#третий корень
  #x0=8
  #x1=6

  xi=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0))
  #print(xi)
  dx=[x1]
  dy=[abs(x1-x0)]

  while abs(xi - x1) > eps:
    
    dx.append(xi)
    dy.append(abs(xi-x1))

    x0 = x1
    x1=xi
    xi=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0))
    
  plt.plot(dx,dy)
  plt.xlabel("x_i")
  plt.ylabel("x_i-x_i-1")
  plt.title("Сходимость")
  #plt.scatter(dx,dy)
    #print(xi)
  return xi
```
