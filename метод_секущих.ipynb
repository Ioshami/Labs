```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return math.log(x)-5*math.cos(x)

a=1.
b=8.
eps=0.01
n=100
```

Вид функции

```python
ff = np.vectorize(f)
x = np.arange (a,b,0.1)
y = ff(x)
plt.plot(x,y)
```

функция для метода секущих

x0>x1 писать

```python

def secantMethod(f,x0,x1,iter):
#первый корень
  #x0 = (a + b) / 2
  #x1=a
#второй корень
  #x0=6
  #x1=4
#третий корень
  #x0=8
  #x1=1

  xi=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0))
  #print(xi)
  while abs(xi - x1) > eps:
    x0 = x1
    x1=xi
    xi=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0))

    #print(xi)
  return xi
```

первый корень

```python
x0 = (a + b) / 2
x1=a
print(secantMethod(f,x0,x1,n))
```

второй корень

```python
x0=6
x1=4
print(secantMethod(f,x0,x1,n))
```

третий корень

```python
x0=8
x1=6
print(secantMethod(f,x0,x1,n))
```
