```python
import numpy as np
import math
```

```python
def f(x):
  return math.log(x)-5*math.cos(x)

a=1
b=8
n=2000001
h = (b - a) / n #шаг
```



---

Первая производная

image.png

```python
def f_firstDer(x0):
  return (f(x0+h)-f(x0))/h
```

```python
f_firstDer(1)
```

Вторая производная

image.png

```python
def f_secondDer(x0):
  return (f(x0+h)-2*f(x0)+f(x0-h))/h**2
```

```python
f_secondDer(1)
```

```python
def simpson(f,a,b,n):
  sum=0.0
  for i in range(1,n-1,2):
    sum+=h/3*(f(a+(i-1)*h)+4*f(a+i*h)+f(a+(i+1)*h))  
  return sum
```

```python
def simpson_ver2(f,a,b,n):
  sum=0.0
  for i in range(1,n-1,2):
    sum+=h/6*(f(a+(i-1)*h)+4*f(a+i*h)+f(a+(i+1)*h))  
  return sum
```
