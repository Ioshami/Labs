Метод Симпсона

```python
import numpy as np
import math
```

```python
def f(x):
  return math.log(x)-5*math.cos(x)

a=1
b=8
n=2000001
h = (b - a) / n #шаг
```

```python
def simpson(f,a,b,n):
  sum=0.0
  for i in range(1,n-1,2):
    sum+=h/3*(f(a+(i-1)*h)+4*f(a+i*h)+f(a+(i+1)*h))  
  return sum
```

```python
print(simpson(f,a,b,n)) 
```
